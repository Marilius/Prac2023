```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv
```

```python
data = {}
with open(r"./research.csv") as csvfile:
    csvreader = csv.reader(csvfile)
    for row in csvreader:
        if row[0] != "series":
            series, iter, mut_prob, time_sec, best_criterion = row
            if int(series) not in data:
                data[int(series)] = {}
            if int(iter) not in data[int(series)]:
                data[int(series)][int(iter)] = {
                    "mut_prob": float(mut_prob),
                    "time_sec": float(time_sec),
                    "best_criterion": int(best_criterion)
                }
```

```python
data
```

```python
plt.figure(figsize=(15, 9))
plt.title("      Pmut")

diff = []
for i in data:
    a = [data[i][j]["best_criterion"] for j in data[i]]
    diff.append(max(a) - min(a))

plt.plot(range(len(data.keys())), diff, 'm')
plt.xticks(range(len(data.keys())))

plt.grid()
plt.show()
```
